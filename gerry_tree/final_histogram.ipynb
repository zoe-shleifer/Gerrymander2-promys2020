```python
import networkx as nx
import json
import matplotlib.ticker as ticker
from helpers import make_graph, make_map,cleanup,signed_color
import networkx as nx
from tqdm.notebook import tqdm
import numpy as np
import pandas as pd
import geopandas as gpd
from random import choice
from termcolor import colored
from scipy import stats
import timeit
import json
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from pathlib import Path
from typing import List, Dict
import pickle
from IPython.display import HTML
import matplotlib.font_manager
from IPython.core.display import HTML
%config InlineBackend.figure_format = 'retina'
import numpy as np
import matplotlib.pyplot as plt

def pickle_load(path):
        with open(path,"rb") as f:
            return pickle.load(f)
results = pickle_load(f'single_chain{2}.pkl')
x = pd.Series(results)
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
p99 = x.quantile(.99)
SCORE = 3.56
cm = plt.cm.get_cmap('plasma')
n, bins, patches = plt.hist(x, 50, density=False,edgecolor='w')
bin_centers = 0.5 * (bins[:-1] + bins[1:])
scale_col = lambda x: 2*(x - min(bin_centers))/max(bin_centers)
col,quant,orig = map(scale_col, (bin_centers,p99,SCORE))
[plt.setp(p, 'facecolor', cm(c)) for c, p in zip(col, patches)]
plt.axvline(p99, color=cm(quant),linestyle=':',label=r'99$^{th}$ percentile')
plt.axvline(SCORE, color=cm(orig),label='Enacted Districts')
plt.legend(loc='upper right',framealpha=1);plt.grid(False)
plt.xlabel('Segregation Score'); plt.ylabel('Number of Maps')
plt.title('Randomly Generated Maps')
plt.show()
```
